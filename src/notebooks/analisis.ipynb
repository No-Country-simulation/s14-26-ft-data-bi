```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ruta = "../../data/resultados/emociones_clase1.csv"
df = pd.read_csv(ruta)
print(df.head())
```

```python
print(df.describe().round(2))
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Emocion', order=df['Emocion'].value_counts().index)
plt.title('Emociones que mas se repite')
plt.xlabel('Emoción')
plt.ylabel('Cantidad de repeticion de emociones')
plt.xticks(rotation=45)
plt.show()
```
